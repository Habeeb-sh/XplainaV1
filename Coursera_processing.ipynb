<h3>2) TF-IDF Skill Importance</h3>

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
# READ DATA
courser_v3 = pd.read_csv("coursera-v3.csv")
```

```python
# 2. CLEAN / PARSE SKILLS
courser_v3["skills_list"] = courser_v3["skills"].apply(
    lambda x: [
        s.strip().strip('"').strip()
        for s in x.strip("{}").split(",")
    ]
)
# Convert list → text so TF-IDF can process it
courser_v3["skills_text"] = courser_v3["skills_list"].apply(lambda x: " ".join(x))
```

```python
# 3. TF-IDF VECTORIZE
vectorizer = TfidfVectorizer(tokenizer=lambda x: x.split(), lowercase=False)
tfidf_matrix = vectorizer.fit_transform(courser_v3["skills_text"])

# Extract vocabulary + scores
feature_names = vectorizer.get_feature_names_out()
tfidf_sums = tfidf_matrix.sum(axis=0).A1  # sum TF-IDF for each skill

# Create a dataframe of (skill, score)
tfidf_scores = pd.DataFrame({
    "skill": feature_names,
    "score": tfidf_sums
})
```

```python
# 4. SELECT TOP 25 TF-IDF SKILLS
top_N = 25
top_skills = (
    tfidf_scores.sort_values("score", ascending=False)
                 .head(top_N)["skill"]
                 .tolist()
)

print("Top 25 TF-IDF Skills:")
print(top_skills)
```

```python
# 5. FILTER SKILLS USING TOP TF-IDF SKILLS
courser_v3["filtered_skills"] = courser_v3["skills_list"].apply(
    lambda skills: [s for s in skills if s in top_skills]
)
```

```python
# 6. CREATE ONE-HOT (0/1) COLUMNS
mlb = MultiLabelBinarizer()
binary_matrix = mlb.fit_transform(courser_v3["filtered_skills"])

binary_df = pd.DataFrame(binary_matrix, columns=mlb.classes_)

```

```python
# 7. MERGE INTO MAIN DF
courser_v3 = pd.concat([courser_v3, binary_df], axis=1)
courser_v3
```

```python
# 8. SAVE FINAL RESULT
courser_v3.to_csv("courses-with-top25-tfidf-skills.csv", index=False)
```

<h3>3) Semantic clustering</h3>

```python
#install the library
#!pip install -U sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import os
```

```python
INPUT_CSV = "coursera-v3.csv"               # change to your csv path
SKILLS_COL = "skills"                     # column with skills like {" SQL"," Python",...}
K = 25                                    # number of representative skills (20-25 recommended)
EMBED_MODEL = "all-MiniLM-L6-v2"          # small+fast; change if you prefer
RANDOM_STATE = 42
OUTPUT_PREFIX = "embedding_clustering"    # prefix for saved files
```

```python
# 1) Read + parse skills
courser_v3 = pd.read_csv(INPUT_CSV)

def parse_skills_cell(cell):
    """Turn: {" SQL"," Python"} -> ['SQL','Python'] (robust)"""
    if pd.isna(cell):
        return []
    s = str(cell).strip()
    if s.startswith("{") and s.endswith("}"):
        s = s[1:-1]
    # split by comma but handle possible comma inside skill (rare)
    parts = [p.strip().strip('"').strip() for p in s.split(",") if p.strip() != ""]
    # remove empty strings and normalize whitespace
    parts = [ " ".join(p.split()) for p in parts if p != "" ]
    return parts

courser_v3["skills_list"] = courser_v3[SKILLS_COL].apply(parse_skills_cell)

# Build unique skill list and frequency
all_skills = courser_v3.explode("skills_list")["skills_list"].dropna()
skill_counts = all_skills.value_counts()
unique_skills = skill_counts.index.tolist()
print(f"Unique skills found: {len(unique_skills)}")
```

```python
# 2) Compute embeddings for unique skills
model = SentenceTransformer(EMBED_MODEL)
# encode in batches for memory safety
batch_size = 128
embeddings = []
for i in range(0, len(unique_skills), batch_size):
    batch = unique_skills[i:i+batch_size]
    emb = model.encode(batch, show_progress_bar=False, normalize_embeddings=True)  # normalized helps cosine
    embeddings.append(emb)
embeddings = np.vstack(embeddings)  # shape (n_skills, dim)
print("Embeddings computed. Shape:", embeddings.shape)
```

```python
# 4) Cluster embeddings with MiniBatchKMeans
k = K
kmeans = MiniBatchKMeans(n_clusters=k, random_state=RANDOM_STATE, batch_size=256)
kmeans.fit(embeddings)
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_  # shape (k, dim)
```

```python
# 5) For each cluster choose representative skill
# Two options: 'most_frequent' OR 'closest_to_centroid'
# We'll choose most_frequent when available, else fallback to closest_to_centroid.

skill_to_cluster = dict(zip(unique_skills, cluster_labels))
cluster_to_skills = {i: [] for i in range(k)}
for skill, cl in skill_to_cluster.items():
    cluster_to_skills[cl].append(skill)

# Precompute mapping skill -> freq
freq_map = skill_counts.to_dict()

representatives = []
for cl in range(k):
    skills_in_cl = cluster_to_skills[cl]
    if len(skills_in_cl) == 0:
        representatives.append(None)
        continue
    # find the most frequent skill in this cluster
    most_freq_skill = max(skills_in_cl, key=lambda s: freq_map.get(s, 0))
    # find closest-to-centroid candidate
    indices = [unique_skills.index(s) for s in skills_in_cl]
    cl_embs = embeddings[indices]
    centroid = centroids[cl]
    # cosine similarity since embeddings were normalized, dot product ~ cosine
    sims = cl_embs.dot(centroid)
    closest_skill = skills_in_cl[int(np.argmax(sims))]
    # choose: prefer most_freq_skill unless it's frequency==1 and there exists a better medium frequency,
    # but generally most_freq_skill is good for explainability. We'll pick most frequent.
    rep_skill = most_freq_skill
    representatives.append({
        "cluster": cl,
        "representative": rep_skill,
        "rep_freq": freq_map.get(rep_skill, 0),
        "closest_skill": closest_skill,
        "cluster_size": len(skills_in_cl)
    })

rep_df = pd.DataFrame(representatives)
print("Representative skills (per cluster):")
print(rep_df.head(30))
```

```python
# 6) Create filtered skills per course keeping only chosen representatives
rep_set = set(rep_df["representative"].dropna().tolist())

def keep_reps(skills):
    return [s for s in (skills or []) if s in rep_set]

courser_v3["filtered_representative_skills"] = courser_v3["skills_list"].apply(keep_reps)
```

```python
# 7) Binarize into 0/1 columns
# We want columns in a stable order: use the representatives in cluster order
rep_order = rep_df["representative"].tolist()
# Clean rep_order remove Nones and duplicates (just in case)
rep_order = [r for r in rep_order if r is not None]
rep_order = list(dict.fromkeys(rep_order))

for rep in rep_order:
    courser_v3[f"skill__{rep}"] = courser_v3["filtered_representative_skills"].apply(lambda sl: 1 if rep in sl else 0)
```

```python
# 8) Save artifacts
os.makedirs("output", exist_ok=True)
rep_df.to_csv(os.path.join("output", f"{OUTPUT_PREFIX}_representatives.csv"), index=False)
# Save mapping skill -> cluster
map_df = pd.DataFrame({
    "skill": unique_skills,
    "cluster": cluster_labels
})
map_df["freq"] = map_df["skill"].map(freq_map)
map_df.to_csv(os.path.join("output", f"{OUTPUT_PREFIX}_skill_clusters.csv"), index=False)
# Save final dataset
courser_v3.to_csv(os.path.join("output", f"{OUTPUT_PREFIX}_courses_with_representatives.csv"), index=False)

print("Saved:")
print(" - representatives ->", os.path.join("output", f"{OUTPUT_PREFIX}_representatives.csv"))
print(" - skill->cluster map ->", os.path.join("output", f"{OUTPUT_PREFIX}_skill_clusters.csv"))
print(" - final dataset ->", os.path.join("output", f"{OUTPUT_PREFIX}_courses_with_representatives.csv"))
```

```python
# 9) Quick coverage check (how many courses have at least one rep)
courser_v3["has_rep"] = courser_v3[[f"skill__{r}" for r in rep_order]].sum(axis=1) > 0
coverage = courser_v3["has_rep"].mean()
print(f"Coverage: {coverage*100:.2f}% of courses contain at least one representative skill")
```
