<h3>Handling the Missing values</h3>

```python
#the null values insights
coursera.isna().sum()
```

```python
#Dropping the null values from skills column
coursera = coursera.dropna(subset=['skills'])
```

```python
coursera
```

```python
#checking the missing value precentage for reveiw count
missing_percent = coursera['review_count'].isna().mean() * 100
print(missing_percent)
```

```python
#If missing % is too high (> 20–30%), mean imputation becomes bad.
if missing_percent < 10:
    print("OK to consider imputation")
else:
    print("Too many NaNs — consider dropping column or advanced imputation")
```

```python
#removing the 'k' from the review_count to numerical values
def convert_k_to_number(x):
    if isinstance(x, str):
        if 'k' in x.lower():
            return float(x.lower().replace('k','')) * 1000
        else:
            return float(x)
    return x
```

```python
coursera = coursera.copy()
```

```python
coursera['review_count'] = coursera['review_count'].apply(convert_k_to_number)
```

```python
skew = coursera['review_count'].skew()
print(skew)
```

```python
coursera
```

```python
#replacing the null/nan values with the mean of the review count column
coursera['review_count'] = coursera['review_count'].fillna(coursera['review_count'].mean())
```

```python
#replacing the null/nan values with the mean of the rating column
coursera['rating'] = coursera['rating'].fillna(coursera['rating'].mean())
```

```python
coursera
```

```python
coursera = coursera.dropna(subset=['level'])
coursera = coursera.dropna(subset=['duration'])
```

```python
#Taking a copy from orginial data set
new_coursera = coursera.copy()
```

```python
new_coursera
```

```python
# Force string
new_coursera['skills'] = new_coursera['skills'].astype(str)

# Remove braces and quotes + normalize commas
new_coursera['skills'] = (
    new_coursera['skills']
        .str.replace(r'[{}"]', '', regex=True)
        .str.replace(r'\s*,\s*', ',', regex=True)
)

# Convert to list
new_coursera['skills'] = new_coursera['skills'].apply(lambda x: x.split(',') if x != 'nan' else [])
```
